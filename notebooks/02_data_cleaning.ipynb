```python
import pandas as pd
```

```python
df_raw = pd.read_csv('../data/raw_reviews_mypertamina.csv')
print (f'Total Data = {len(df_raw)}')
```

```python
df_clean = df_raw.dropna(subset=['Reviews']).copy()
```

```python
df_clean['Jumlah_Kata'] = df_clean['Reviews'].apply(lambda x : len(str(x).split()))
df_clean = df_clean[df_clean['Jumlah_Kata'] >= 5]
```

```python
df_clean = df_clean.drop_duplicates(subset=['Reviews'])
```

```python
df_clean = df_clean.drop(columns = ['Jumlah_Kata'])
print (f'Total Cleaned Data = {len(df_clean)}')
```

```python
df_clean.to_csv('../data/cleaned_reviews_mypertamina.csv',index = False)
```
