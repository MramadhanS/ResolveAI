```python
import os
from dotenv import load_dotenv
from llama_index.core import SimpleDirectoryReader, VectorStoreIndex, Settings
from llama_index.llms.gemini import Gemini
from llama_index.embeddings.gemini import GeminiEmbedding

print("Menyiapkan Sistem RAG...")

# 1. Muat API Key dari file .env
load_dotenv()
api_key = os.getenv("GEMINI_API_KEY")

# 2. Atur Model LLM dan Embedding Sesuai Daftar Resmi API
Settings.llm = Gemini(model="models/gemini-2.5-flash", api_key=api_key)
Settings.embed_model = GeminiEmbedding(model_name="models/gemini-embedding-2", api_key=api_key)

# 3. Baca semua dokumen Markdown di dalam folder knowledge_base
documents = SimpleDirectoryReader("../data/knowledge_base/").load_data()
print(f"Berhasil memuat {len(documents)} blok dokumen dari folder.")

# 4. Ubah teks menjadi Vektor dan simpan di memori
index = VectorStoreIndex.from_documents(documents)
print("Vector Database berhasil dibangun!")

# 5. Ubah index menjadi mesin pencari (Query Engine)
query_engine = index.as_query_engine()
```

```python
# Skenario Pengujian 1: Syarat Pendaftaran LPG 3 Kg
pertanyaan_1 = "Saya mau mendaftar sebagai Usaha Mikro untuk beli LPG 3 Kg. Selain KTP, dokumen apa lagi yang wajib saya bawa ke Pangkalan?"

print(f"Pertanyaan 1: {pertanyaan_1}\n")
print("ResolveAI sedang mencari jawaban di SOP...\n")

jawaban_1 = query_engine.query(pertanyaan_1)
print("--- Jawaban ---")
print(jawaban_1.response)
print("\n" + "="*50 + "\n")

# Skenario Pengujian 2: Kebijakan Privasi
pertanyaan_2 = "Saya agak takut data saya disalahgunakan. Apakah boleh KTP dan KK saya difotokopi lalu disimpan saja oleh petugas Pangkalannya?"

print(f"Pertanyaan 2: {pertanyaan_2}\n")
print("ResolveAI sedang mencari jawaban di Kebijakan Privasi...\n")

jawaban_2 = query_engine.query(pertanyaan_2)
print("--- Jawaban ---")
print(jawaban_2.response)
```
