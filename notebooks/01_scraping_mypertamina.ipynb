```python
import pandas as pd

from google_play_scraper import Sort,reviews

print ('Tools Ready')
```

```python
app_id = "com.dafturn.mypertamina"

result, continuation_token = reviews(
    app_id,
    lang='id',             
    country='id',          
    sort=Sort.NEWEST,      
    count=5000,            
    filter_score_with=1    
)

print(f"Done ! {len(result)} Data.")
```

```python
df_reviews = pd.DataFrame(result)

df_reviews = df_reviews [['userName','at','score','content']]

df_reviews.columns = ['Username','Waktu','Rating','Reviews']

df_reviews.head()
```

```python
file_path = '../data/raw_reviews_mypertamina.csv'

df_reviews.to_csv(file_path, index=False)
```
